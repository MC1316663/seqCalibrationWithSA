### 주의사항: 같은 놈이 두번 들어오는 경우도 있음. 이런 경우는? 1,000초 이상 차이나면 다른 것으로 합시다. 나중에 분포보고 다시 또 설정하면 되니깐.

```python
d1_ex.sort_values(by = ['id', 'time'], ascending = True, inplace = True)
d2_ex.sort_values(by = ['id', 'time'], ascending = True, inplace = True)
d3_ex.sort_values(by = ['id', 'time'], ascending = True, inplace = True)

d1_ex.reset_index(inplace = True)
d2_ex.reset_index(inplace = True)
d3_ex.reset_index(inplace = True)

d1_ex
```

```python
def timeComplete(df):
    
    eachId = 0
    ids = []
    sid = []
    time = []
    tem_time = 0
    visit_time = 1
    visit = []

    for i in tqdm(range(len(df))):
        if eachId != df['id'][i]: #eachId랑 다르면 새로운 놈이 들어온 것
            ids.append(df['id'][i])
            sid.append(df['sid'][i])
            time.append(df['time'][i])
            eachId = df['id'][i]  #each Id 초기화
            tem_time = df['time'][i] #tem_time초기화
            
            visit_time = 1
            visit.append(visit_time)

        else: #같은 id에 다른 타임.
            timeGap = df['time'][i] - tem_time #compute time gap
            if timeGap == 0: #error
                None
            elif timeGap == 1:  #1이면 그냥 저장하고 다음으로.
                ids.append(df['id'][i])
                sid.append(df['sid'][i])
                time.append(df['time'][i])
                visit.append(visit_time)
            elif timeGap < 1000: #차이가 1초 이상 있네. 그런데 1000초 이하네
                for j in range(timeGap):   #만약 갭이 2라면 0,1
                    ids.append(df['id'][i])
                    sid.append(df['sid'][i])
                    time.append(df['time'][i] - timeGap + j + 1) #갭 채우기
                    visit.append(visit_time)

                tem_time = df['time'][i] #tem_time초기화

            else: #차이가 1000초 이상이네 -> 여기 할 차례. 내 생각엔 여기에 '횟수' 컬럼도 넣어서 한번 더면 번호 매기자. 나중에 observed data만들 때 필요.
                ids.append(df['id'][i])
                sid.append(df['sid'][i])
                time.append(df['time'][i])
                tem_time = df['time'][i] #tem_time초기화
                visit_time = visit_time + 1 #1 번 더 방문
                visit.append(visit_time)
    
    result = pd.DataFrame(
                {'id' : ids,
                 'sid' : sid,
                 'time' : time,
                 'visit' : visit}
                )
    
    return result
                    
```

```python
d1_timeComp = timeComplete(d1_ex)
d2_timeComp = timeComplete(d2_ex)
d3_timeComp = timeComplete(d3_ex)
```

```python
'''데이터의 분포가 정규 분포를 이루지 않거나 한 쪽으로 skewed 한 경우, 데이터의 IQR 값을 이용해 이상치를 탐지'''
'''이사분범위를 바탕으로 Q3 + 1.5*IQR 이상, Q1–1.5*IQR 이하의 값을 outlier로 정의: IQR rule'''
## IQR 구하기
num = d1_sum['sec']

Q1 = num.quantile(.25)
Q3 = num.quantile(.75)
IQR = Q3-Q1

upRange = Q3 + 1.5*IQR
downRange = Q1 - 1.5*IQR
if downRange < 0:
    downRange = 0
print("outlier 제거이전 머문 시간의 평균: ")
print("outlier: <", downRange, ", > ", upRange)
```

```python
## outlier 제거
def outlier(day, df):
    series = df['sec']
    Q1 = series.quantile(.25)
    Q3 = series.quantile(.75)
    IQR = Q3-Q1

    upRange = Q3 + 1.5*IQR
    downRange = Q1 - 1.5*IQR
    if downRange < 0:
        downRange = 0
        
    print("outlier 제거이전 day: ", day, "의 머문 시간의 평균: ", np.mean(series), ", 표준편차: ", np.std(series))
    print("outlier: <", downRange, ", > ", upRange)
    
    #outlier 제거
    df = df[df['sec'] < upRange]
    series = df['sec']
    print("outlier 제거이후 day: ", day, "의 머문 시간의 평균: ", np.mean(series), ", 표준편차: ", np.std(series))
    
    return df
```

```python
d1_remove_outlier = outlier(1, d1_sum)
d2_remove_outlier = outlier(1, d2_sum)
d3_remove_outlier = outlier(1, d3_sum)
```

```python
#저장
d1_remove_outlier['day'] = 1
d2_remove_outlier['day'] = 2
d3_remove_outlier['day'] = 3

path = 'E:/Work_since_2022/2022_Indoor_contact_simulation/1_Equifinality_Paper/Data/Observed_Data/'
d1_remove_outlier.to_csv(path + "d1_remove_outlier.csv")
d2_remove_outlier.to_csv(path + "d2_remove_outlier.csv")
d3_remove_outlier.to_csv(path + "d3_remove_outlier.csv")
```

```python
# https://medium.com/the-researchers-guide/finding-the-best-distribution-that-fits-your-data-using-pythons-fitter-library-319a5a0972e9
import seaborn as sns
from fitter import Fitter, get_common_distributions, get_distributions

data = d1_remove_outlier['sec'].values
f = Fitter(data,
          distributions= ['gamma',
                         'lognorm',
                         'beta',
                         "norm",
                         'rayleigh'])
f.fit()
f.summary()
```

```python
'''이제 이 값이 확정된 분포'''
f.get_best(method = 'sumsquare_error')
```

```python
shape, scale = f.fitted_param["gamma"][0], f.fitted_param["gamma"][2]
shape, scale
```

```python
'''이제 이 분포에서 뽑아보기'''
data = np.random.gamma(shape, scale, 100) #100개 뽑기

import matplotlib.pyplot as plt
import scipy.special as sps  
count, bins, ignored = plt.hist(data, 50, density=True)
y = bins**(shape-1)*(np.exp(-bins/scale) /  
                      (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')  
plt.show()
```
